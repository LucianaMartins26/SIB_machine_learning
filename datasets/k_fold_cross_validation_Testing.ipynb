```python
import numpy as np
from SIB_machine_learning.src.si.model_selection.cross_validation import k_fold_cross_validation, \
    leave_one_out_cross_validation
from SIB_machine_learning.src.si.io.csv_file import read_csv
from SIB_machine_learning.src.si.models.logistic_regression import LogisticRegression

dataset_path = 'breast//breast-bin.csv'
dataset = read_csv(dataset_path, sep=',', features=True, label=True)

logistic_model = LogisticRegression()

scores_k_fold = k_fold_cross_validation(logistic_model, dataset, cv=5, seed=1)

print("k-fold Cross-Validation Scores:")
print(scores_k_fold)

print(f'Mean score: {np.mean(scores_k_fold)} +/- {np.std(scores_k_fold)}')

scores_loocv = leave_one_out_cross_validation(logistic_model, dataset, seed=1)

print("\nLeave-One-Out Cross-Validation Scores:")
print(scores_loocv)

print(f'Mean score: {np.mean(scores_loocv)} +/- {np.std(scores_loocv)}')
```
